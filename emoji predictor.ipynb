# **step-3  getting glove vector**

## glove vectors-made by stanford 

```python
f=open(r"C:\Users\Sahil Garg\Desktop\machine_learning new\words embeddings\odd one out challenge\glove.6B.50d.txt",encoding='utf-8')
```

```python
embedding_index={}
for line in f:
    values=line.split()
    word=values[0]
    coef=np.asarray(values[1:],dtype='float32')
    embedding_index[word]=coef
f.close()
```

```python
embedding_index["eat"]
```

```python
emb_dim=embedding_index["eat"].shape   ## so each word is represented by a 50 dimensional vector
```

# step 4 - converting sentence into vectors(Embedding layer output)

```python
def embedding_output(X):
    maxlen=10
    embedding_out= np.zeros((X.shape[0],10,50))
    
    
    for ix in range(X.shape[0]):
        X[ix]=X[ix].split()    ## so that "i am sleeping" can divide into "i","am","sleeping" ## tokenization
        
        for ij in range(len(X[ix])):
            
            # go to every word in the current (ix) sentence
            
            try:
                embedding_out[ix][ij]= embedding_index[X[ix][ij].lower()]
            except:
                embedding_out[ix][ij]= np.zeros((50,))
                
    return embedding_out
```

```python
embedding_matrix_train=embedding_output(X_train)
embedding_matrix_test=embedding_output(X_test)
```

```python
embedding_matrix_test.shape
```

```python
embedding_matrix_train.shape
```

## step-5 :- Define the RNN/LSTM model

```python
from keras.models import Sequential
from keras.layers import *
```

```python
model=Sequential()
model.add(LSTM(64,input_shape=(10,50))) ## 64 is the shape of activation vector
model.add(Dropout(0.5))  ## to do regularization
model.add(Dense(5))    ## 5 neuron becoz output has 5 classes
model.add(Activation("softmax"))
model.compile(loss="categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
from keras.callbacks import EarlyStopping ,ModelCheckpoint

Model_stop=ModelCheckpoint("best_model.h5",monitor='val_loss',verbose=True,save_best_only=True)
early_stop=EarlyStopping(monitor='val_acc',patience=10)
```

```python
model.fit(embedding_matrix_train,Y_train,batch_size=64,epochs=100,shuffle=True,validation_split=0.2,callbacks=[Model_stop,early_stop])
```

```python
model.load_weights("best_model.h5")
```

```python
pred=model.predict_classes(embedding_matrix_test)
```

```python
model.evaluate(embedding_matrix_test,Y_test)
```

## stacked LSTM to improve Accuracy

```python
from keras.models import Sequential
from keras.layers import *
```

```python
model=Sequential()
model.add(LSTM(64,input_shape=(10,50),return_sequences=True)) ## 64 is the shape of activation vector
model.add(Dropout(0.5))  ## to do regularization
model.add(LSTM(64,return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(5))    ## 5 neuron becoz output has 5 classes
model.add(Activation("softmax"))
model.compile(loss="categorical_crossentropy",optimizer='rmsprop',metrics=['accuracy'])
model.summary()
```

```python
from keras.callbacks import EarlyStopping ,ModelCheckpoint

Model_stop=ModelCheckpoint("best_model.h5",monitor='val_loss',verbose=True,save_best_only=True)
early_stop=EarlyStopping(monitor='val_acc',patience=10)
model.fit(embedding_matrix_train,Y_train,batch_size=64,epochs=100,shuffle=True,validation_split=0.2,callbacks=[Model_stop,early_stop])
```

```python
model.load_weights("best_model.h5")
```

```python
model.evaluate(embedding_matrix_test,Y_test)
```
